```python
cols_to_keep = ['latin_name', 'polish_name', 'english_name', 'type', 'invasiveness']


plants_df = plants_df[cols_to_keep]

plants_df
```

```python
# 1. Definiujemy punktację
score_rules = {
    'bardzo inwazyjny gatunek obcy': 5,
    'średnio inwazyjny gatunek obcy': 4,
    'mało inwazyjny gatunek obcy': 3,
    'nieinwazyjny gatunek obcy': 2
}

# 2. Zamieniamy tekst na punkty
plants_df['points'] = plants_df['invasiveness'].map(score_rules).fillna(0).astype(int)

# 3. Sprawdzamy wynik (wyświetlamy nazwę i punkty)
plants_df
```
